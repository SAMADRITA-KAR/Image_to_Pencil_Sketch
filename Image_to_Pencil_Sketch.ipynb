```python
#Importing libraries
import cv2
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading and Plotting the image
i=cv2.imread("Portrait.jpg")
i=cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(i)
a=cv2.waitKey(0)
d=cv2.destroyAllWindows()
plt.axis("off")
plt.title("Original")
plt.show()
```

```python
#Grayscaling the image
ig=cv2.cvtColor(i,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(6,6))
plt.imshow(ig,cmap="gray")
plt.axis("off")
plt.title("Grayscale")
plt.show()
```

```python
#Inverting the image
im=255 - ig
plt.figure(figsize=(6,6))
plt.imshow(im,cmap="gray")
plt.axis("off")
plt.title("Inverted")
plt.show()
```

```python
#Smoothening the image
bi=cv2.GaussianBlur(im,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(6,6))
plt.imshow(im,cmap="gray")
plt.axis("off")
plt.title("Smoothened")
plt.show()
```

```python
#Converting to pencil sketch
f=cv2.divide(ig,255 - bi,scale=255)
plt.figure(figsize=(6,6))
plt.imshow(f,cmap="gray")
plt.axis("off")
plt.title("Final Sketch")
plt.show()
```
